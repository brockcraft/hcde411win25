## Exercise 1: Introductory Interactions
In the code cell below, generate a scatter plot of the min vs. max temp from the `seattle_weather` data set. Your scatter plot should enable panning and zooming, as well as include tooltips that show the type of weather. You should generate the same plot twice, once using the manual selection syntax and once using the shorthand syntax for zooming and panning. Display them side by side (the code for this is given to you).

```python
from vega_datasets import data
weather = data.seattle_weather()

manual = alt.Chart(weather).mark_circle().add_selection(
    alt.selection_interval(bind='scales')
).encode(
    alt.X('temp_min:Q'),
    alt.Y('temp_max:Q', axis=alt.Axis(minExtent=30)),
    tooltip=['weather:N']
)

shorthand = alt.Chart(weather).mark_circle().encode(
    alt.X('temp_min:Q'),
    alt.Y('temp_max:Q', axis=alt.Axis(minExtent=30)),
    tooltip=['weather:N']
).interactive()

# Uncomment the below when finished
manual | shorthand
```

## Exercise 3: Brushing and Linking

Generate two linked charts below: 1) a bar chart which shows the change in precipitation over the years and 2) a scatter plot of min vs. max temp. They should be linked such that brushing (via clicking and dragging) on the bar chart should filter out the points on the scatter plot so that only the selected ones appear (the filtered out points shouldn't just be faded out; they should be completely invisible). The code to concatenate the final two charts is given to you.

```python
brush = alt.selection_interval(
    encodings=['x']
)

precip = alt.Chart(weather).mark_bar().encode(
    alt.X('date:T'),
    alt.Y('precipitation:Q')
).add_selection(
    brush
)

temps = alt.Chart(weather).mark_circle().encode(
    alt.X('temp_min:Q'),
    alt.Y('temp_max:Q', axis=alt.Axis(minExtent=30)),
    opacity=alt.condition(brush, alt.value(0.75), alt.value(0.0))
)

# Uncomment after completing the above
alt.vconcat(precip, temps).properties(spacing=5)
```
